### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 3. Assign it to a variable called euro12.

```python
euro12 = pd.read_csv('Euro_2012_stats_TEAM.csv', sep=',')
euro12
```

### Step 4. Select only the Goal column.

```python
euro12['Goals']
```

### Step 5. How many team participated in the Euro2012?

```python
# 행 개수만 가져오기
euro12.shape[0]
```

### Step 6. What is the number of columns in the dataset?

```python
# 열 개수만 가져오기
euro12.shape[1]
```

### Step 7. View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline

```python
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
discipline
```

### Step 8. Sort the teams by Red Cards, then to Yellow Cards

```python
discipline.sort_values(['Red Cards', 'Yellow Cards'], ascending=False)
```

### Step 9. Calculate the mean Yellow Cards given per Team

```python
round(discipline['Yellow Cards'].mean())
```

### Step 10. Filter teams that scored more than 6 goals

```python
euro12[euro12.Goals > 6]
```

### Step 11. Select the teams that start with G

```python
# Team 열이 포함된 DataFrame에서
# 각 팀 이름이 'G'로 시작하는지 확인하여 True/False 시리즈 생성
euro12.Team.str.startswith('G')

# 위에서 생성된 Boolean 시리즈를 이용해 True인 행만 필터링
euro12[euro12.Team.str.startswith('G')]
```

### Step 12. Select the first 7 columns

```python
euro12.iloc[:, 0:7]
```

### Step 13. Select all columns except the last 3.

```python
euro12.iloc[:, :-3]
```

### Step 14. Present only the Shooting Accuracy from England, Italy and Russia

```python
euro12.loc[euro12.Team.isin(['England', 'Italy', 'Russia']), ['Team', 'Shooting Accuracy']]
```

### Iloc (integer location) and loc (location)
* Use `.loc` when you know the labels (names) of the rows/columns.
* Use `iloc` when you know the integer position of the rows/columns.

[Iloc vs Loc in Pandas: A Guide With Examples](https://www.datacamp.com/tutorial/loc-vs-iloc)

```python
euro12.head(10)
```

```python
# Selecting a single row using .loc
euro12.loc[0]
```

```python
# Conditional selection of rows using .loc
euro12[euro12['Team'] != 'England']
```

```python
# Selecting a single column using .loc
euro12.loc[:, 'Team']
```

```python
euro12.loc[:, ['Team', 'Goals']]
```

```python
# Selecting a slice of columns using .loc
euro12.loc[:, 'Team':'Shooting Accuracy']
```

```python
# Combined row and column selection using .loc
euro12.loc[euro12['Goals'] > 5, 'Team':'Goals']
```

```python
# Selecting a single row using .iloc
euro12.iloc[1]
```

```python
# Selecting multiple rows using .iloc
euro12.iloc[[0, 3, 4]]
```

```python
# Selecting a slice of rows using .iloc
euro12.iloc[1:4]
```

```python
# Selecting a single column using .iloc
euro12.iloc[:, 4]
```

```python
# Selecting multiple columns using .iloc
euro12.iloc[:, [0, 3]]
```

```python
# Selecting a slice of columns using .iloc
euro12.iloc[:, 1:3]
```

```python
euro12.iloc[[0, 1, 4], 1:4]
```
